## サンプルプログラムを実行
- matplotlibにて簡単なデータの可視化

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams["font.family"] = "IPAexGothic"

data = pd.read_csv('COVID-19.csv', usecols=['年代'], dtype={'年代': object})
age_datas = data['年代']

fix_age_datas = [] # 年齢のリスト
infected_list = [] # 感染者数のリスト

# 空白を削除
for age_data in age_datas:
    x = age_data.replace(' ',"")
    fix_age_datas.append(x)

# 年齢のリストをソート
age_list = sorted(set(fix_age_datas))

# 感染者数をリストに追加
for i in range(0,len(age_list)):
    infected_list.append(fix_age_datas.count(age_list[i]))

# 棒グラフ描画
plt.figure(figsize=(20, 10), dpi=50)
plt.rcParams["font.size"] = 25
plt.title("COVID-19年齢別感染者数 (2020/4/8時点)")
plt.bar(age_list,infected_list)
```
